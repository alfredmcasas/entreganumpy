```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Solo cargamos las columnas 7 , 8 , 9 correspondientes a la abreviación del pais, el año y el valor del gasto publico
gobiernosdf = np.genfromtxt("D:\Documentos\Master\R\datos\government.csv", delimiter = ",",skip_header = 1,usecols = (7,8,9),dtype = str)
```

```python
#El dataframe solo tiene 3 entradas para el año 2021 y todos los demas valores son n.a por lo que para facilitar la creación
# de series elimino los datos de 2021

#Con esta funcion sacamos los indices donde el año sea 2021

indices_2021 = np.where(gobiernosdf[:,1] == "2021")[0]

# Y borramos las filas donde se encuentren esos valores

gobiernosdf =  np.delete(gobiernosdf, indices_2021, axis=0)
gobiernosdf
```

```python
# Para comprobar agrupamos por pais, contando la cantidad de siglas diferentes y agrupandolas. Vemos que ahora si hay 9 
# entradas por país distinto del 2012 al 2020.

paises = gobiernosdf[:,0]
pais , indice = np.unique(paises, return_inverse=True)

agrupacion_pais = [paises[indice == i] for i in range(len(pais))]
agrupacion_pais

conteo = [len(unico) for unico in agrupacion_pais]
conteo
```

```python
# Creo un array unico solo con los valores y transformamos de str a float para poder operar.

valores = gobiernosdf[:,2]
valores = valores.astype(float)
valores
```

```python
# Con este bucle pretendo calcular la media de cada pais y meter el resultado en una lista, al tener 9 observaciones por pais
# se facilita el hacerlo. No tiene pinta de ser muy eficiente esto pero no se me ha ocurrido nada más.

medias_pais = []
i = 0
suma = []
for entry in valores:
    if i < 8:
        suma.append(entry)
        i += 1
    else:
        suma.append(entry)
        media = sum(suma)/9
        medias_pais.append(media)
        i=0
        suma = []
       
```

```python
# Le adjuntamos las siglas del pais al lado con la funcion zip y ya tenemos una lista con el gasto publico medio en % del PIB por país desde 2012 hasta 2020.

media_pais = list(zip(medias_pais, pais))
media_pais
```

```python
# Con la lista anterior guardamos en una variable el valor que equivale a la media de gasto en los 27 paises de la eurozona.
Media_eurozona = list(filter(lambda x: x[1]=='EU27_2020',media_eurozona))
Media_eurozona = Media_eurozona[0][0]

Media_eurozona
```

```python
# Y sacamos una lista de los paises que esten por encima de esa media

Paises_media_superior = list(filter(lambda x: x[0] >Media_eurozona,media_pais))
Paises_media_superior
```

Vemos que hay 12 paises con un gasto medio entre 2012 y 2020 por encima del gasto general de los paises de la eurozona

```python
# Y por debajo

Paises_media_inferior = list(filter(lambda x: x[0] < Media_eurozona, media_pais))
Paises_media_inferior
```
